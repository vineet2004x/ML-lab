```python
X= df.drop(columns=['isPlay'])
X
```

```python
y=df["isPlay"]
y
```

```python
Xn=np.array(X)
```

```python
print(Xn)
```

```python
yn=np.array(y)
```

```python
print(yn)
```

```python
def train(X,y):
  for i,val in enumerate(y):
    if val=='yes':
      h=X[i].copy()
      break
  for i,val in enumerate(X):
    if y[i]=='yes':
      for z in range(len(h)):
        if val[z] !=h[z]:
          h[z]='?'
        else:
          pass
  return h
```

```python
print(train(Xn,yn))
```
